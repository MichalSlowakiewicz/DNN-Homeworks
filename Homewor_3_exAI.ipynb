*   **Partial Dependence**

In this point, we aim to obtain Paritial Dependence (PD) for variable $X_1$. We will do that by directly computing from definition presented during lecture:

$$
g_1^{PD}(z) = \mathbb{E}_{X_2}(f(z, X_2)) = \mathbb{E}_{X_2}((z+X_2)^2) = \mathbb{E}_{X_2}(z^2+2zX_2+X_2^2)=\mathbb{E}_{X_2}(z^2)+\mathbb{E}_{X_2}(2zX_2)+\mathbb{E}_{X_2}(X_2^2) = z^2 + 0 + \frac{1}{3} = z^2 + \frac{1}{3}
$$

In the last step, we used two equalities, which can be easily obtained:



1.   $\mathbb{E}_{X_2}X_2 = 0$ since $X_2$ ~ $U[-1,1]$
2.   $\mathbb{E}_{X_2}X_2^2 = \int_{-1}^1 x^2 \frac{1}{2}dx = \frac{1}{3}$


*   **Marginal Effects**

We will calculate the value of Marginal Effects (ME) for variable $X_1$. Analogously, the direct computation from definition will be coducted:

$$
g_1^{ME}(z) = \mathbb{E}_{X_2 | X_1 = z}(f(z, X_2)) = \mathbb{E}_{X_2 | X_1 = z}((z+X_2)^2) = (*)
$$

Now, it is time to use the additional information from this task - $X_1$ and $X_2$ are fully dependent. Because of that property, $X_1 = z$ implies that $X_2$. We may continue our computations with new knowledge:

$$
(*) = \mathbb{E}_{X_2 | X_1 = z}((z+z)^2) = (2z)^2 = 4z^2
$$


*   **Accumulated Local Effects**

Finally, we may move to computing Accumulated Local Effects (ALE) for variable $X_1$. Once again, our calculations will follow from the definition:

$$
g_1^{ALE}(z) = \int_{z_0}^z \mathbb{E}_{X_2 | X_1 = y}\frac{\partial f(x_1, x_2)}{\partial x_1} dy = \int_{-1}^z \mathbb{E}_{X_2 | X_1 = y}\frac{\partial (x_1+x_2)^2}{\partial x_1} dy = \int_{-1}^z \mathbb{E}_{X_2 | X_1 = y} (2x_1+2x_2) dy = (*)
$$

Analogoulsy, as in computations for ME, we will take advantage of full dependency of $X_1$ and $X_2$ are fully dependent ($X_1 = y ⇒ X_2 = y$):

$$
(*) = \int_{-1}^z \mathbb{E}_{X_2 | X_1 = y} (4x_1) dy = \int_{-1}^z (4y) dy = 2y^2 |^z_{-1} = 2z^2 - 2
$$
